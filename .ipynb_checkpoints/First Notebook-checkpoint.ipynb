#### Before we dive in PyRevit, I would like to introduce to you a little about Revit API. <font color=purple>API</font> stands for <font color=purple>Application Programming Interface,</font> Revit is a complex 3D building information modeling software and it's written in C Sharp (C#) language. Together API is an interface between third party developers and the software itself. Keep in mind the third developers can't create all the tools of functionality that you see in the software but only the stuffs that the developers of this application have decided to expose to the outside world to use and modified.  

#### Third party developer like us create <font color = blue >PyRevit</font> inside Revit enviroment.
